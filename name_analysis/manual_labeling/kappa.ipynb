```python
import os
import pandas as pd
from sklearn.metrics import cohen_kappa_score
print(os.listdir('/content/drive/MyDrive/Model hubs/23Summer - PTM Naming Convention/Data/Outliers'))
DIR_HEAD = '/content/drive/MyDrive/Model hubs/23Summer - PTM Naming Convention/Data/Outliers'
```

```python
ex1 = pd.read_excel('/content/drive/MyDrive/Model hubs/23Summer - PTM Naming Convention/Data/Outliers/Researcher1 (Aaryan)/external_outliers_researcher1.xlsx')
ex2 = pd.read_excel('/content/drive/MyDrive/Model hubs/23Summer - PTM Naming Convention/Data/Outliers/Researcher2 (Shiva)/external_outliers_researcher2.xlsx')
combined = pd.concat([ex1['Real Outlier?'], ex2['Real Outlier?']], axis=1, keys=['col1', 'col2'])

combined['col1'] = combined['col1'].str.lower()
combined['col2'] = combined['col2'].str.lower()
combined.dropna(inplace=True)
cohen_kappa_score(combined['col1'], combined['col2'])
```

```python
ex_arr1 = []
ex_arr2 = []
for col in ex1.keys():
    # print(col)
    cnt = 0
    if col == "Real Outlier?":
      # if col == "Issue #": continue
      for idx, (i, j) in enumerate(zip(ex1[col], ex2[col])):
          # print(i, j)

          if pd.isna(i) or pd.isna(j): continue
          cnt += 1
          ex_arr1.append(str(i).lower())
          ex_arr2.append(str(j).lower())
cohen_kappa_score(ex_arr1, ex_arr2)
```

```python
in1 = pd.read_excel('/content/drive/MyDrive/Model hubs/23Summer - PTM Naming Convention/Data/Outliers/Researcher1 (Aaryan)/internal_outliers_researcher1.xlsx')
in2 = pd.read_excel('/content/drive/MyDrive/Model hubs/23Summer - PTM Naming Convention/Data/Outliers/Researcher2 (Shiva)/internal_outliers_researcher2.xlsx')
in_arr1 = []
in_arr2 = []
for col in in1.keys():
    # print(col)
    cnt = 0
    if col == "Real Outlier?":
      # if col == "Issue #": continue
      for idx, (i, j) in enumerate(zip(in1[col], in2[col])):
          # print(i, j)

          if pd.isna(i) or pd.isna(j): continue
          cnt += 1
          in_arr1.append(str(i).lower())
          in_arr2.append(str(j).lower())
cohen_kappa_score(in_arr1, in_arr2)
```

```python
gpt = pd.read_excel('/content/drive/MyDrive/Model hubs/23Summer - PTM Naming Convention/Data/FSE Data/GPT Evaluation.xlsx')
gpt_data = gpt.loc[:,'Semantic Pattern 1':'Semantic Pattern 7'].fillna(' ').astype(str).apply(' '.join, axis=1)
manual_data = gpt.loc[:,'Manual labled 1':'Manual labled 7'].fillna(' ').astype(str).apply(' '.join, axis=1)
cohen_kappa_score(gpt_data, manual_data)
```

```python
# Initialize counter for different rows
count_diff = 0

# Iterate through the lists and compare values
for val1, val2 in zip(gpt_data, manual_data):
    if val1 != val2:
        count_diff += 1

# Calculate the percentage of differences
total_rows = len(gpt_data)  # or len(manual_data), they should be the same length
percentage_diff = (count_diff / total_rows) * 100

print(f"Percentage of differences: {percentage_diff}%")
```
