```python
import re


def camelCaseChecker(txt) -> bool:
    return re.search("^[a-z]+([A-Z][a-z]+)+$", txt)
    # doesn't take into account "HellohiThere". If first word is capitalized, it's false

def snake_case_checker(txt) -> bool:
    return re.search("^[A-Za-z0-9]+(_[A-Za-z0-9]+)*$", txt)

def kebab_case_checker(txt) -> bool:
    return re.search("^[A-Za-z0-9]+(-[A-Za-z0-9]+)*$", txt)

def mixed_caseChecker(txt) -> set:
    matches = re.findall("(?:(?:[a-z]+_[a-z]+)|([a-z]+-[a-z]+)|([a-z]+(?:[A-Z][a-z]*)+))", txt)
    return matches

x = camelCaseChecker("helloThere_chicken")
y = mixed_caseChecker("hello_There-chicken")
if len(set(y)) >= 2:
    print("gay")
if x:
    print("gay2")
else:
    print ("straight")

import json

f = open("final_cleaned_output.json")

data = json.load(f)

for i in data:
    print(i)

from matplotlib import pyplot as plt


names = ["snake_case", "kebab-case", "camelCase", "mixEd-Case"]

numbers = [100, 48, 122, 303]

plt.bar(names, numbers)
plt.xlabel("Naming styles")
plt.ylabel("Frequencies")
plt.title("Frequencies of Naming Styles")
plt.show()
```
